```python
from matplotlib import pyplot as plt    
from matplotlib import style
import seaborn as sns
style.use('ggplot')
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import dendrogram, linkage,  fcluster
#min simlarity
from scipy.cluster.hierarchy import single
#max simlarity
from scipy.cluster.hierarchy import complete
#group average similarity
from scipy.cluster.hierarchy import average
#centroid similarity
from scipy.cluster.hierarchy import centroid
```

```python
#import data
df = pd.read_csv('B.txt',header=None, delimiter=' ')
df.columns = ['x', 'y']
#change dataframe to numpy arrary
df = df.values
df
```

```python
#distance
y = pdist(df)
```

```python
Z=linkage(df,method="single", metric='euclidean')
dendrogram(Z)
plt.figure(figsize=(10,10))
plt.show()
clusters=fcluster(Z, 2, 'maxclust')
plt.figure(figsize=(10,10))
plt.scatter(df[:,0], df[:,1], c=clusters, cmap='prism')
plt.show()
```

```python
Z=linkage(df,method="complete", metric='euclidean')
dendrogram(Z)
plt.figure(figsize=(10,10))
plt.show()
clusters=fcluster(Z, 2, 'maxclust')
plt.figure(figsize=(10,10))
plt.scatter(df[:,0], df[:,1], c=clusters, cmap='prism')
plt.show()
```

```python
Z=linkage(df,method="average", metric='euclidean')
dendrogram(Z)
plt.figure(figsize=(10,10))
plt.show()
clusters=fcluster(Z, 2, 'maxclust')
plt.figure(figsize=(10,10))
plt.scatter(df[:,0], df[:,1], c=clusters, cmap='prism')
plt.show()
```

```python
Z=linkage(df,method="centroid", metric='euclidean')
dendrogram(Z)
plt.figure(figsize=(10,10))
plt.show()
clusters=fcluster(Z, 2, 'maxclust')
plt.figure(figsize=(10,10))
plt.scatter(df[:,0], df[:,1], c=clusters, cmap='prism')
plt.show()
```
